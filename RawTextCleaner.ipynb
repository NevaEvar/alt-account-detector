```python
import csv 
import re


#readPath = 'v1/content/drive/Shareddrives/Datasets/cleaned-v4/discord-v1-detox-antispam/[136542963336478720] [part 1].txt'

def RawTextToCSV(readPath):
    
    with open(readPath, 'r', encoding='utf-8') as readFile:
        rawText = readFile.read()

        #makes the text continuous and breaks it down to entries based on tabs
        #also separates ?, !, ., , and @
        rawText = rawText.replace('\\n', ' ¤ ')
        rawText = rawText.replace('\n', ' ¤ ')
        rawText = re.sub(r'([A-z0-9])([.?!,@])', r'\1 \2', rawText)
        rawText = re.sub(r'([@])([A-z0-9])', r'\1 \2', rawText)
        lines = rawText.split('\t')

        # Extracts user and message 
        dialogue = []
        formula = re.compile(r'^(.*?):\s*(.*)')

        for line in lines:
            match = formula.match(line.strip())
    
            if match:
                user = match.group(1).strip()
                message = match.group(2).strip()
                dialogue.append([user, message])
        
        with open('CleanText.csv', 'a', newline='', encoding='utf-8') as writeFile:
            writer = csv.writer(writeFile)
            writer.writerows(dialogue)

```
